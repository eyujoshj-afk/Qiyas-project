```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Task 1: Explore the Dataset
# Task 1: Explore the Dataset using the absolute path
movies = pd.read_csv(r"C:\Users\Pc\Desktop\Ai\movies_updated.csv.csv")
```

```python
print(movies.head())
```

```python
print(movies.head())
print(movies.info())
print(movies.describe())
```

```python
#Task 2: Number of Movies Released Each Year
movies_year = movies.groupby("year").size()
plt.figure(figsize=(8, 5))
plt.plot(
    movies_year.index,
    movies_year.values,
    marker="s",
    color="red",
    linestyle="--",label='Number of movie'
)
plt.title("Number of Movies Released Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.legend()
plt.show()
```

```python

#Task 3: Top 10 Movie Genres
top_genres = movies["genre"].value_counts().head(10).sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(top_genres.index, top_genres.values, color="yellow", edgecolor="black")
ax.set_xlabel("Genre")
ax.set_ylabel("Number of Movies")
ax.set_title("Top 10 Movie Genres")
plt.show()
```

```python
#Task 4: Top 10 Directors
top_directors = (
    movies["director"].value_counts().head(10).sort_values(ascending=True))

fig, ax = plt.subplots(figsize=(8, 5))
ax.barh(top_directors.index, top_directors.values, color="purple", edgecolor="black")
ax.set_xlabel("Number of Movies")
ax.set_ylabel("Director")
ax.set_title("Top 10 Directors by Number of Movies")

plt.tight_layout()
plt.show()
```

```python
#Task 5: Distribution of Movie Ratings
plt.figure(figsize=(10, 5))
plt.hist(
    movies["score"].dropna(),
    bins=20,
    color="gold",
   # edgecolor="black",
    alpha=0.7,
)
plt.title("Distribution of Movie Ratings (Scores)")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#Task 6: Average Movie Score by Genre
avg_score_genre = movies.groupby("genre")["score"].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(8, 5))
avg_score_genre.plot(kind="bar", color="blue", edgecolor="black")
plt.title("Average Movie Score by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Score")
#plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Task 7: Top 10 Highest Grossing Movies
top_grossing = movies.nlargest(10, "gross")[["name", "gross"]]
plt.figure(figsize=(10, 6))
plt.barh(
    top_grossing["name"][::-1],
    top_grossing["gross"][::-1] / 1e9,
    color="black",
    edgecolor="black",
)
plt.title("Top 10 Highest Grossing Movies")
plt.xlabel("Gross Revenue (in Billions USD)")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```
